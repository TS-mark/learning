```python
from PIL import ImageFont, ImageDraw, Image
import numpy as np
from paddleocr import PaddleOCR, draw_ocr
import time
import cv2
r'''
Using paddleOCR module，recognize Chinese(including numbers and characters)

INFO
----
Date:2022-11-07
Author:zhang.wenzhao1

PARAMETERS
----
img_path: origin picture path
save_path:save picture path

RETURN
----

USING
----
case1:
OCR_using_paddle(r"C:\jinxData\TT100K\data\test\28040.jpg")
case2:
OCRusingPaddle(r"C:\Users\Mark\Pictures\tm.png", "demo2.jpg")

DEPENDINGS
----
PIL
numpy
paddleocr == 2.6.1.0
paddlepaddle
cv2
'''

def OCR_using_paddle(img_path, save_path = None):

    
    old_time = time.time()
    ocr = PaddleOCR(lang="ch")  # 首次执行会自动下载模型文件，可以通过修改 lang 参数切换语种
    time_mark2 = time.time()
    result = ocr.ocr(img_path)

    point = None
    image = cv2.imread(img_path)
    img= Image.fromarray(image)  # 转换为PIL库可以处理的图片形式
    draw=ImageDraw.Draw(img)
    for line in result:
        if len(line) <= 1:
            break
        aa=line[1][0]
        for i in range(len(line)):
            point = line[i]
            draw.rectangle([int(point[0][0][0]),int(point[0][0][1]), int(point[0][2][0]),int(point[0][2][1])]\
                                                                       ,outline = (0,255,0))
            x_c = int(point[0][1][0])-int(point[0][0][0])
            y_c = int(point[0][0][1])-int(point[0][2][1])
    #        计算下字体的大小,我的计算方式：检测框长度/字符数
            font1=ImageFont.truetype("./simsun.ttc",min(int(x_c/len(point[1][0])),-y_c))  # 设置字体类型和大小，注意：simsun.ttc为字体库文件需要你的程序目录下有此文件，如果没有需要获取，获取方法在win10系统中：在我的电脑>C盘>windows>Fonts的文件夹下面复制任意一个一个中文的库到你程序的目录下即可，不复制获取库的路径可能也可以没试过
    #       openCV不支持中文，这里用PIL画图
            draw.rectangle([int(point[0][0][0]),int(point[0][0][1])+y_c, \
                            int(point[0][2][0]),int(point[0][2][1])+y_c],fill = (0,0,0))
            draw.text((point[0][0][0],point[0][0][1]+y_c),point[1][0], \
                      font=font1,fill=(200,200,200))  # 图片上加入中文，使用与cv2.putText类似

    #         cv2.rectangle(image, (int(point[0][0][0]),int(point[0][0][1])+y_c), \
    #                       (int(point[0][2][0]),int(point[0][2][1])+y_c), (255, 0, 0), 1)
    #         cv2.putText(image, str(point[1][0]), (int(point[0][0][0]),int(point[0][0][1])),\
    #                     cv2.FONT_HERSHEY_SIMPLEX, (point[0][1][0]-point[0][0][0])/len(point[1][0])/50, (0,0,0),2)

    new_time=time.time()
    print('using time：', new_time-old_time)

    img1= np.array(img)  # 转换为PIL库可以处理的图片形式
    cv2.namedWindow("demo1", cv2.WINDOW_NORMAL)
    if save_path:
        cv2.imwrite(save_path,img1)
    cv2.imshow("demo1",img1)
    key = cv2.waitKey(0)
    return result

```
