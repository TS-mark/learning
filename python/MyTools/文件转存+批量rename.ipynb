```python
import os
import shutil
```

```python
ori_path = r"Z:\datasets\returned"
dir_path = "G:/datasets/ggge/"
if not os.path.exists(dir_path):
    os.makedirs(dir_path)

file_end = ".jpg"
# 命名方式 name_started + file_name
name_started = "rt"
file_name = 0


file_list = os.listdir(ori_path)
for f in file_list:
    if not f.endswith(file_end):
        continue
    while( os.path.exists(
        os.path.join(dir_path, 
                    name_started+str(file_name) + file_end))):
        file_name += 1
    shutil.move(os.path.join(ori_path, f), os.path.join(dir_path, name_started+str(file_name) + file_end))
    print("pic {} has saved to path {}".format(os.path.join(ori_path, f), os.path.join(dir_path, name_started+str(file_name) + file_end)))
    file_name += 1
    
        


```
