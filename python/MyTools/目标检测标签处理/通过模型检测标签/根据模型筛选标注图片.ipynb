```python
import os
# 准备：已经训练好的模型，通过模型训练得到图像标注推理的结果
#
```

```python
det_res_dir1 = r"E:\yolov5-7.0\runs\detect\exp10\labels"
det_list1 = os.listdir(det_res_dir1)
det_res_dir2 = r"E:\yolov5-7.0\runs\detect\exp13\labels"
det_list2 = os.listdir(det_res_dir2)
res_file = []
```

```python
len(det_list1)
```

```python
det_list = list(set(det_list1) & set(det_list2))
len(det_list)
```

```python
forward_file = ""
consecutive = 0
now_series = -22
cnt_series = 0
for det_res in det_list:
    # print(det_res)
    # print(det_res.split(".")[0].split("_")[-1])
    temp_series = int(det_res.split(".")[0].split("_")[-1]) 

    # consecutive = consecutive + 1 if now_series + 1 == temp_series else 0
    if det_res.split(".")[0].split("_")[:-1] != forward_file.split(".")[0].split("_")[:-1]:
        consecutive = 0
    if now_series + 1 == temp_series:
        consecutive += 1
        cnt_series = 0
    else:
        if consecutive > 2:
            cnt_series = 2
        for i in range(cnt_series):

            save_path = "_".join(forward_file.split(".")[0].split("_")[:-1]) + "_" + str(now_series+1+i) + ".jpg"
            res_file.append(save_path)
            print("额外保存", save_path)
        consecutive = 0
    res_file.append(det_res.replace(".txt",".jpg"))
    now_series = temp_series
    forward_file = det_res
```

```python
len(res_file)
```

```python
res_file = list(set(res_file))
len(res_file)
```

```python
bad_filepath = []
```

```python
from shutil import copyfile
for f in res_file:
    cp_path = os.path.join("G:/to_be_detected/to_be_detected",f)
    saved_path = os.path.join("G:/to_annotation",f)
    if os.path.exists(cp_path):
        copyfile(cp_path, saved_path)
    else:
        bad_filepath.append(f)
    
```
