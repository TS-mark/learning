```python
"""从A文件夹下摘取与B文件夹里内容相同的文件转存到C文件中"""
import os
import shutil

# 源文件夹A和目标文件夹C的路径
folder_a_path = r'G:\TLD_train7\images\test'

folder_b_path = r'E:\yolov5-7.0\runs\val\exp115\totrain'
folder_c_path = r'G:\TLD_train7\images\train'

# 获取文件夹B中的文件名列表
folder_b_files = os.listdir(folder_b_path)

# 遍历文件夹A中的文件
for filename in os.listdir(folder_a_path):
    file_a_path = os.path.join(folder_a_path, filename)
    
    # 检查文件是否与文件夹B中的文件重名
    if filename in folder_b_files:
        # 构建目标文件路径
        file_c_path = os.path.join(folder_c_path, filename)
        
        # 移动文件到文件夹C
        shutil.move(file_a_path, file_c_path)
        print(f"Moved {filename} to folder C")

print("Done")

folder_a_label_path = r'G:\TLD_train7\labels\test'
folder_c_label_path = r'G:\TLD_train7\labels\train'


# 遍历文件夹A_label中的文件
for filename in os.listdir(folder_a_label_path):
    file_a_path = os.path.join(folder_a_label_path, filename)
    
    # 检查文件是否与文件夹B中的文件重名
    if filename.replace(".txt",".jpg") in folder_b_files:
        # 构建目标文件路径
        file_c_path = os.path.join(folder_c_label_path, filename)
        
        # 移动文件到文件夹C
        shutil.move(file_a_path, file_c_path)
        print(f"Moved {filename} to folder C")

print("Done")

```
