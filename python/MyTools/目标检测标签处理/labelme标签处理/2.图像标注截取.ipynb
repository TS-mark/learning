```python
r'''
capture the picture for classification task

INFO
----
Date:2022-11-17
Author:zhang.wenzhao1

PARAMETERS
----
img_path: (str)origin picture path
ann_path: (str)annotation path
save_path: (str)save picture path
datasets: (list)including capture datasets

RETURN
----
None

USING
----
consult example

DEPENDINGS
----
cv2
'''
def capture_pic(pic_path, ann_path, save_path, datasets = ["train", "test"]):
    abn = []
    for d in datasets:
        pic_list = os.listdir(os.path.join(pic_path, d))
        for p in pic_list:
            img = cv2.imread(pic_path + r"/"+ d +"/" + p)
            #判断文件尾缀，end.jpg
            if p.endswith(".jpg"):
                #标签文件是否存在
                if os.path.exists(ann_path + r"/"+ d +"/" + p.replace(".jpg", ".txt")):
                    with open(ann_path + r"/"+ d +"/" + p.replace(".jpg", ".txt")) as f:
                        cnt = 0
                        for i in f.readlines():
                            l = i.split(" ")

                            res = img[max(int((float(l[2])-0.5*float(l[4]))*img.shape[0]), 0):min(int((float(l[2])+0.5*float(l[4]))*img.shape[0]),img.shape[0]),\
                                       max(int((float(l[1])-0.5*float(l[3]))*img.shape[1]), 0):min(int((float(l[1])+0.5*float(l[3]))*img.shape[1]),img.shape[1])]
                            
                            # print(p,max(int((float(l[2])-0.5*float(l[4]))*480), 0),min(int((float(l[2])+0.5*float(l[4]))*480),480),\
                            #      max(int((float(l[1])-0.5*float(l[3]))*640), 0),min(int((float(l[1])+0.5*float(l[3]))*640),640))
                            # 错误标注场景：w、h等于0
                            if max(int((float(l[2])-0.5*float(l[4]))*480), 0) == min(int((float(l[2])+0.5*float(l[4]))*480),480)\
                                or max(int((float(l[1])-0.5*float(l[3]))*640), 0) == min(int((float(l[1])+0.5*float(l[3]))*640),640):
                                continue
                            cnt += 1
                            class_path = save_path + r"/" + l[0]
                            if not os.path.exists(class_path):
                                os.makedirs(class_path)
                            cv2.imwrite(class_path + r"/" + "{}".format(p.replace(".jpg","___{}.jpg".format(cnt))), res)
                else:
                    abn.append(d + "\t" + p)
                    
    print("存在未标注图片")
    print(abn)
```
