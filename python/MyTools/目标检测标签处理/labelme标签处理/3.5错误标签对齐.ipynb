```python
cls_path = "g:\\datasets\\classifier\\classifier"
ori_path = r"g:\datasets\toPicked"



"""
pic_name: 例：2022830_23-49-32_0_685___2.jpg
label_path: 例：F:\datasets\toPicked\labels
cls: 上面那个cls，每个类别的文件夹
"""
def alter_label(pic_name, label_path, cls):
    lb_path = os.path.join(label_path, pic_name.split("__")[0] + ".txt")
    line = int(pic_name.split("___")[1].split(".")[0])
    # print(line)
    file_context = ""
    if not os.path.exists(lb_path):
        print("label {} not exists".format(lb_path))
        return
    with open(lb_path,"r") as f:
        for i in f:
            if line == 1:
                i = cls + " " + " ".join(i.split(" ")[1:])
            line -= 1
            file_context += i
                 
    with open(lb_path,"w") as f:
        f.write(file_context)


""" 
将摘选出来的图片还原到对应标签
"""
def ClassPath_get(cls_path, ori_path):
    cls = os.listdir(cls_path)
    for c in cls:
        label_box = os.path.join(cls_path,c)
        for lb in os.listdir(label_box):
            alter_label(lb, ori_path, c)

ClassPath_get(cls_path, os.path.join(ori_path, "labels"))
```
